```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
pokemon = pd.read_csv("Pokedex_Ver_SV1.csv")
pokemon.head()
```

```python
pokemon = pokemon.drop(["Original_Name", "Name", "Generation", "Ability1", "Ability2", "Ability_Hidden", "Color"], axis = 1)
```

```python
pokemon.head()
```

```python
pokemon = pokemon.drop(["Gender_Male", "Gender_Female", "Gender_Unknown", "Egg_Steps", "Egg_Group1", "Egg_Group2", "Region_Form"], axis = 1)
```

```python
pokemon = pokemon.drop(["E_HP", "E_Attack", "E_Defense", "E_SP_Attack", "E_Speed", "E_SP_Defense"], axis = 1)
pokemon.head()
```

```python
np.random.seed(3)

pokemon.fillna("No", inplace = True)

pokemon1 = pd.get_dummies(pokemon[["Type1", "Type2", "Mega_Evolution_Flag"]], )
```

```python
pokemon1
```

```python
pokemon.head()
```

```python
pokemon = pd.concat([pokemon.drop(["Type1", "Type2", "Mega_Evolution_Flag"], axis=1), pokemon1], axis=1)
pokemon
```

```python
pokemon["Height"] = pokemon["Height"].str.replace("m", "")
pokemon["Weight"] = pokemon["Weight"].str.replace("kg", "")
pokemon
```

```python
X = pokemon.drop("Category", axis = 1)
y = pokemon["Category"]

X_train, X_test, y_train, y_test = train_test_split(X, y)

clf = RandomForestClassifier()

clf.fit(X_train, y_train)

y_preds = clf.predict(X_test)

clf.score(X_test, y_test)
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_preds))

# Classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_preds))
```
